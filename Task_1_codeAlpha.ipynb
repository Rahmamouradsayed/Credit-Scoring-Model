```python
# import libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report , confusion_matrix ,accuracy_score
from sklearn.linear_model import LogisticRegression
```

```python
#load data_set
data =pd.read_csv('Credit_Card_dataset.csv')
```

```python
data.shape
```

```python
data.head()
```

```python
dataset = data.drop('ID', axis=1)
```

```python
# to check from Null Values
dataset.isna().sum()
```

```python
# Fill missing values in the dataset using backward fill
dataset=dataset.fillna(method='bfill')
```

```python
#check all null values are removed after replace null values with backward fill
dataset.isna().sum()
```

```python
data.describe()
```

```python
dataset.shape
```

```python
label_encoder = LabelEncoder()

dataset['CODE_GENDER']=label_encoder.fit_transform(dataset['CODE_GENDER'])
dataset['FLAG_OWN_CAR']=label_encoder.fit_transform(dataset['FLAG_OWN_CAR'])
dataset['FLAG_OWN_REALTY']=label_encoder.fit_transform(dataset['FLAG_OWN_REALTY'])
dataset['NAME_INCOME_TYPE'] = label_encoder.fit_transform(dataset['NAME_INCOME_TYPE'])
dataset['NAME_EDUCATION_TYPE'] = label_encoder.fit_transform(dataset['NAME_EDUCATION_TYPE'])
dataset['NAME_FAMILY_STATUS'] = label_encoder.fit_transform(dataset['NAME_FAMILY_STATUS'])
dataset['NAME_HOUSING_TYPE'] = label_encoder.fit_transform(dataset['NAME_HOUSING_TYPE'])
dataset['OCCUPATION_TYPE'] = label_encoder.fit_transform(dataset['OCCUPATION_TYPE'])
```

```python
dataset
```

```python
# Extract the target variable (Y) from the dataset
Y=dataset.iloc[:,0].values

# Extract the feature variables (X) from the dataset
X=dataset.iloc[:, 1:16].values
```

```python
# Split the data into training and testing sets
X_train ,X_test , Y_train , Y_test=train_test_split(X,Y , test_size=0.2,random_state=0,stratify=Y)
```

```python
# Create an instance of StandardScaler
scaler = StandardScaler()
```

```python
# Standardize the feature variables in the training set
X_train = scaler.fit_transform(X_train)
```

```python
# Standardize the feature variables in the testing set
X_test = scaler.transform(X_test)
```

```python
# Create an instance of the LogisticRegression classifier
classifier=LogisticRegression()
```

```python
# Fit the classifier to the training data
classifier.fit(X_train,Y_train)
```

```python
# Make predictions on the testing set
y_pred = classifier.predict(X_test)
```

```python
print(confusion_matrix(Y_test,y_pred))
```

```python
print(accuracy_score(Y_test, y_pred))
```

```python
predictions = classifier.predict_proba(X_test)
predictions
```

```python
df_prediction_prob = pd.DataFrame(predictions, columns = ['prob_0', 'prob_1'])
df_prediction_target = pd.DataFrame(classifier.predict(X_test), columns = ['predicted_TARGET'])
df_test_dataset = pd.DataFrame(Y_test,columns= ['Actual Outcome'])

dfx=pd.concat([df_test_dataset, df_prediction_prob, df_prediction_target], axis=1)

dfx.to_csv("Credit_Card_dataset.csv", sep=',', encoding='UTF-8')

dfx.head()
```
